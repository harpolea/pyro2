```python
from __future__ import print_function

import numpy as np

import mesh.patch as patch
import multigrid.MG as MG
```

We want to solve $\phi_{xx} + \phi_{yy} = -2[(1-6x^2)y^2(1-y^2) + (1-6y^2)x^2(1-x^2)]$ on $[0,1]\times[0,1]$ with homogeneous Dirichlet boundary conditions (this example comes from "A Multigrid Tutorial").

```python
nx = ny = 256
mg = MG.CellCenterMG2d(nx, ny, 
                       xl_BC_type="dirichlet", xr_BC_type="dirichlet",
                       yl_BC_type="dirichlet", yr_BC_type="dirichlet")
```

```python
def rhs(x, y):
    return -2.0*((1.0-6.0*x**2)*y**2*(1.0-y**2) + (1.0-6.0*y**2)*x**2*(1.0-x**2))

mg.init_RHS(rhs(mg.x2d, mg.y2d))
mg.init_zeros()
```

```python
mg.solve()
```

```python
phi = mg.get_solution()
```

```python
pylab.imshow(np.transpose(phi.v()), origin="lower")
```

```python
gx, gy = mg.get_solution_gradient()
```

```python
pylab.subplot(121)
pylab.imshow(np.transpose(gx.v()), origin="lower")
pylab.subplot(122)
pylab.imshow(np.transpose(gy.v()), origin="lower")
```
