# Testing against pyro

Plots based on initial data in Rezzolla & Zanotti 02

```python
from r3d2 import State, RiemannProblem, eos_defns, wave
from IPython.display import display, display_png, display_svg
from matplotlib.pyplot import savefig
import numpy as np
import matplotlib # these two lines set the display so it works
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
eos = eos_defns.eos_gamma_law(5.0/3.0)
eos_reactive = eos_defns.eos_gamma_law_react(5.0/3.0, 0.1, 1.0, 1.0, eos)

U_burnt = State(1.0, 0.5, 0.0, 1.6, eos)
U_reactive = State(0.125, 0.0, 0.0, 1.2, eos_reactive)
rp = RiemannProblem(U_burnt, U_reactive)

display(rp)
display_png(rp)
```

With a tangential velocity, the rarefaction becomes a shock:

```python
U_burnt = State(1.0, 0.5, 0.0, 1.6, eos)
U_reactive = State(0.125, 0.0, 0.9, 1.2, eos_reactive)
rp = RiemannProblem(U_burnt, U_reactive)

display(rp)
display_png(rp)
```

```python
eos = eos_defns.eos_gamma_law(5.0/3.0)
test_1_w_left = State(10.0, 0.0, 0.0, 2.0, eos, label="L")
test_1_w_right = State(2.0, 0.0, 0.0, 0.5, eos, label="R")
test_1_rp = RiemannProblem(test_1_w_left, test_1_w_right)
test_2_w_left = State(1.0, 0.0, 0.0, 10.0, eos, label="L")
test_2_w_right = State(1.0, 0.0, 0.0, 1.0, eos, label="R")
test_2_rp = RiemannProblem(test_2_w_left, test_2_w_right)
test_3_w_left = State(1.0, 0.0, 0.0, 10, eos, label="L")
test_3_w_right = State(1.0, 0.0, 0.9, 1.0, eos, label="R")
test_3_rp = RiemannProblem(test_3_w_left, test_3_w_right)
test_4_w_left = State(1.0, 0.0, 0.9, 10.0, eos, label="L")
test_4_w_right = State(1.0, 0.0, 0.9, 1.0, eos, label="R")
test_4_rp = RiemannProblem(test_4_w_left, test_4_w_right)
test_8_w_left = State(1.0, 0.0, 0.999, 10.0, eos, label="L")
test_8_w_right = State(1.0, 0.0, 0.999, 1.0, eos, label="R")
test_8_rp = RiemannProblem(test_8_w_left, test_8_w_right);
```

```python
for n, i in enumerate([test_1_rp, test_2_rp, test_3_rp, test_4_rp, test_8_rp]):
    display(i)
    display_png(i)
    fig_data = i._figure_data('png')
    figname = '../Writing/figures/rp' + str(n+1) + '.png'
    file = open(figname, 'wb')
    file.write(fig_data)
    file.close()
```

```python
eos = eos_defns.eos_gamma_law(5.0/3.0)
eos_reactive = eos_defns.eos_gamma_law_react(5.0/3.0, 0.1, 1.0, 1.0, eos)

# a single detonation wave
test_5_U_reactive = State(5.0, 0.0, 0.0, 2.0, eos_reactive)
test_5_U_burnt = State(8.113665227084942, -0.34940431910454606, 0.0, 
                2.7730993786742353, eos)
test_5_rp = RiemannProblem(test_5_U_reactive, test_5_U_burnt)

# a single deflagration wave
test_6_U_reactive = State(5.0, 0.0, 0.0, 2.1, eos_reactive)
test_6_U_burnt = State(0.5, 0.7, 0.0, 2.0, eos)
test_6_rp = RiemannProblem(test_6_U_reactive, test_6_U_burnt)

# thing from p-v plots doc file
test_7_U_reactive = State(0.5, 0.0, 0.0, 1.0, eos_reactive)
test_7_U_burnt = State(0.24316548798524526, -0.39922932397353039, 0.0,
                    0.61686385086179807, eos)
test_7_rp = RiemannProblem(test_7_U_burnt, test_7_U_reactive)
```

```python
for n, i in enumerate([test_5_rp, test_6_rp, test_7_rp]):
    display(i)
    display_png(i)
    fig_data = i._figure_data('png')
    figname = '../Writing/figures/rp' + str(n+5) + '.png'
    file = open(figname, 'wb')
    file.write(fig_data)
    file.close()
```

```python
xi = [-1.05]
data2 = test_2_rp.state_l.state()
for wave in test_2_rp.waves:
    xi_wave, data_wave = wave.plotting_data()
   
    xi = np.hstack((xi, xi_wave))
    data2 = np.vstack((data2, data_wave))
xi = np.hstack((xi, [1.05]))
data2 = np.vstack((data2, test_2_rp.state_r.state()))

data3 = test_3_rp.state_l.state()
for wave in test_3_rp.waves:
    xi_wave, data_wave = wave.plotting_data()
    data3 = np.vstack((data3, data_wave))
data3 = np.vstack((data3, test_3_rp.state_r.state()))

data4 = test_4_rp.state_l.state()
for wave in test_4_rp.waves:
    xi_wave, data_wave = wave.plotting_data()
    data4 = np.vstack((data4, data_wave))
data4 = np.vstack((data4, test_4_rp.state_r.state()))

rho2 = data2[:, 0]
p2 = data2[:, 4]
v2 = data2[:, 1]
vt2 = data2[:, 2]

rho3 = data3[:, 0]
p3 = data3[:, 4]
v3 = data3[:, 1]
vt3 = data3[:, 2]

rho4 = data4[:, 0]
p4 = data4[:, 4]
v4 = data4[:, 1]
vt4 = data4[:, 2]
```
